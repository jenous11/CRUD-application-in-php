```python
# now genarating a function

import re

def clean_text(data):

  # checking for url links
  result = re.sub(r'https?[\S\s]+', " ", data)
  partiallycleaned_result=result
  partiallycleaned_result

  #checking for new @ and other mentions
  fully_cleaned_result=re.sub(r'@[a-zA-Z0-9]+'," ",partiallycleaned_result)
  fully_cleaned_result

# #remove special characters
  fully_cleaned_result = re.sub(r'\W+'," ",fully_cleaned_result)
  cleaned_data= fully_cleaned_result.strip().lower()
  return cleaned_data

# output:"you are so stupid check"


#passing data to the function
# data="@ShreyaBafna3 You are SO stupid!! Check http://somelink.com \xc2\xa0"
# clean_text(data)
```

```python
kaggle_clean['Text'] = kaggle_clean['Text'].apply(clean_text)
```
